```python
X = B_C_data.drop(['diagnosis'], axis=1)
X = X.drop(['Unnamed: 32'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Xs = sc.fit_transform(X)
fig, ax = plt.subplots(1, figsize=(20, 8))
sns.boxplot(data=Xs, ax = ax)
```

```python
from sklearn.decomposition import PCA
```

```python

feature_names = list(X.columns)
```

```python
pca = PCA(n_components=10)
Xs_pca = pca.fit_transform(Xs)
```

```python
PCA_df = pd.DataFrame() 
PCA_df['PCA_1'] = Xs_pca[:,0] 
PCA_df['PCA_2'] = Xs_pca[:,1]
```

```python
plt.figure(figsize=(6,6))
```

```python
plt.plot(PCA_df['PCA_1'][B_C_data['diagnosis'] == 'M'],PCA_df['PCA_2'][B_C_data['diagnosis'] == 'M'],'ro', alpha = 0.7, markeredgecolor = 'k')
plt.plot(PCA_df['PCA_1'][B_C_data['diagnosis'] == 'B'],PCA_df['PCA_2'][B_C_data['diagnosis'] == 'B'],'bo', alpha = 0.7, markeredgecolor = 'k')
plt.xlabel('PCA_1') 
plt.ylabel('PCA_2') 
plt.legend(['Malignant','Benign']);
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
B_C_data['diagnosis'] = le.fit_transform(B_C_data['diagnosis'])
B_C_data.head()
y = B_C_data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xs, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.svm import SVC
clf = SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', probability=True)
clf.fit(X_train, y_train)
classifier_score = clf.score(X_test, y_test)
print('The classifier accuracy score is {:03.2f}'.format(classifier_score))
```

```python
from sklearn.model_selection import cross_val_score
n_folds = 5
clf_cv = SVC(C=1.0, kernel='rbf', degree=3, gamma='auto') 
cv_error = np.average(cross_val_score(clf_cv, Xs, y, cv=n_folds)) 
print('The {}-fold cross-validation accuracy score for this classifier is {:.2f}'.format(n_folds, cv_error))
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
clf_fs_cv = Pipeline([
('feature_selector', SelectKBest(f_classif, k=3)),
('svc', SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', probability=True))
])
scores = cross_val_score(clf_fs_cv, Xs, y, cv=5)
```

```python
print(scores)
avg = (100 * np.mean(scores), 100 * np.std(scores)/np.sqrt(scores.shape[0])) 
print("Average score and standard deviation: (%.2f +- %.3f)%%" %avg)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred = clf.fit(X_train, y_train).predict(X_test) 
cm = confusion_matrix(y_test, y_pred)
```
